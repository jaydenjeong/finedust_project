```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# 한국 / 중국의 사회경제 데이터 읽기
pm10 = pd.read_csv("./pm10_final_v4.csv", index_col = 0, header = 0)
pm25 = pd.read_csv("./pm25_final_v4.csv", index_col = 0, header = 0)
```

```python
# 2개의 제거 리스트 읽기
kr_remove = pd.read_csv("./kr_remove.csv", index_col = 0)
cn_remove = pd.read_csv("./cn_remove.csv", index_col = 0)
```

```python
# 제거 리스트 병합
remove_list = pd.concat([kr_remove, cn_remove], axis = 0).reset_index(drop = True)
```

```python
# 제거 리스트 확인. 60개의 피처가 제거 대상임을 확인
remove_list.loc[:, "0"]
```

```python
# 피처 제거
pm10_new = pm10.drop(remove_list.loc[:, "0"], axis = 1)
pm25_new = pm25.drop(remove_list.loc[:, "0"], axis = 1)
```

```python
# 저장
pm10_new.to_csv("./pm10_final_v5.csv")
pm25_new.to_csv("./pm25_final_v5.csv")
```
