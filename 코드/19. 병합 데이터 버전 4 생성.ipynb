```python
import pandas as pd
import numpy as np
```

```python
# 미세먼지 데이터 읽기
pm10 = pd.read_csv("./pm10_final_v1.csv", index_col = 0)
pm25 = pd.read_csv("./pm25_final_v1.csv", index_col = 0)
```

```python
# 미세먼지 데이터의 인덱스값 데이트 타입으로 변환
pm10.index = pd.to_datetime(pm10.index)
pm25.index = pd.to_datetime(pm25.index)
```

```python
# 한국 / 중국의 사회경제 데이터 읽기
kr_economy = pd.read_csv("./독립변수/한국경제지표종합.csv", index_col = 0)
cn_economy = pd.read_csv("./독립변수/중국사회경제종합.csv", index_col = 0)
```

```python
# 경제 데이터의 인덱스값 데이트 타입으로 변환
kr_economy.index = pd.to_datetime(kr_economy.index)
cn_economy.index = pd.to_datetime(cn_economy.index)
```

```python
# 3시간 단위의 시간 리스트 생성
time_list = pd.date_range(start = "2015-01-01-00:00", end = "2023-12-31-21:00", freq = "3H")
```

```python
# 3시간 단위의 데이터프레임 생성
time_df = pd.DataFrame(time_list)
time_df.columns = ["일시"]
time_df = time_df.set_index("일시")
time_df.head()
```

```python
# time_df에 cn_economy와 동일한 컬럼들을 생성, 값은 nan 입력
for column in cn_economy.columns:
    time_df[column] = np.nan
```

```python
time_df.shape
```

```python
cn_economy.shape
```

```python
# 날짜가 일치하는 경우에 cn_economy 값을 입력
for date in cn_economy.index:
    time_df.loc[date, :] = cn_economy.loc[date, :].values
```

```python
cn_economy2 = time_df
```

```python
# 미세먼지, 한국/중국 사회경제 데이터 통합 적용
pm10_v4 = pd.concat([pm10, kr_economy, cn_economy2], axis = 1)
pm25_v4 = pd.concat([pm25, kr_economy, cn_economy2], axis = 1)
```

```python
# 24년도 이후의 데이터 제거
pm10_v4 = pm10_v4[pm10_v4.index < "2024-01-01"]
pm25_v4 = pm25_v4[pm25_v4.index < "2024-01-01"]
```

```python
# 초미세먼지의 경우 18년 이전의 데이터 제거
pm25_v4 = pm25_v4[pm25_v4.index >= "2019-01-01" ]
```

```python
# 첫 데이터 제거
pm10_v4 = pm10_v4.iloc[1:, :]
pm25_v4 = pm25_v4.iloc[1:, :]
```

```python
# csv 파일로 저장
pm10_v4.to_csv("./pm10_final_v4.csv")
pm25_v4.to_csv("./pm25_final_v4.csv")
```
