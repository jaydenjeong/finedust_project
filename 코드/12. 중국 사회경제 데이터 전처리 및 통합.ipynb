- 각각의 데이터에 결측은 없으나 다음의 문제점이 제기된다
- 주가 데이터는 일 단위, 나머지 데이터는 월단위로 이루어져 있다
    - 월 단위의 데이터를 일 단위로 변환할 필요가 있음
- 월 단위의 데이터들은 24년 5월의 데이터가 없다
    - 방법이없음

### 월단위 데이터 통합 후 일단위로 데이터 변경

```python
# 데이터 통합
cn_economy = pd.concat([cn_indus, cn_energ, cn_trans], axis = 1)
cn_economy.head()
```

```python
# 인덱스를 날짜 형식으로 변경
cn_economy.index = pd.to_datetime(cn_economy.index)
cn_economy.head()
```

```python
# 15년 1월 1일부터 24년 4월 30일까지의 날짜 리스트 생성
date_list = pd.date_range(start = "2015-01-01", end = "2024-04-30")
```

```python
# 데이터에 없는 날짜 추가
for date in date_list:
    if date not in cn_economy.index:
        cn_economy.loc[date, :] = np.nan
```

```python
cn_economy.tail()
```

```python
# 인덱스를 기준으로 정렬
cn_economy = cn_economy.sort_index()
```

```python
# ffill을 활용하여 월별로 동일한 데이터 입력
cn_economy = cn_economy.ffill()
```

### 주가 데이터와 병합 및 마무리

```python
# cn_stock의 인덱스를 날짜 타입으로 변경
cn_stock.index = pd.to_datetime(cn_stock.index)
```

```python
# 주가 데이터와의 병합
cn_economy = pd.concat([cn_economy, cn_stock], axis = 1)
```

```python
# 결측치가 있는 5월 데이터 제거
cn_economy = cn_economy.dropna()
```

```python
# 저장 후 확인
cn_economy.to_csv("./중국사회경제/중국사회경제종합.csv")
cn_economy2 = pd.read_csv("./중국사회경제/중국사회경제종합.csv", index_col = 0)
cn_economy2.head()
```
