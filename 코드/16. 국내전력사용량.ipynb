```python
import pandas as pd
import numpy as np
import glob
```

```python
# 국내 전력 사용량 데이터 전부 불러오기
# 불러올 csv 파일들이 있는 베이스 경로 설정
directory = 'C:/Users/Jayden Jeong/OneDrive/바탕 화면/project/final_project/분석/data/국내/전력사용량/'

# 디렉토리 내의 모든 xls 파일 매칭하기
xls_files = glob.glob(directory + '*.xls')

# 모든 xls 파일을 데이터프레임으로 읽어오기
dfs = []
for xls_file in xls_files:
    df = pd.read_excel(xls_file, skiprows = 132)
    dfs.append(df)

# 하나의 데이터프레임으로 병합
total_df = pd.concat(dfs, ignore_index=True)

total_df.head()
```

```python
# 불필요한 컬럼 제거
total_df = total_df.drop("시군구", axis = 1)
```

```python
# "5호미만제거"는 삭제해도될듯함
total_df[total_df["고객호수(호)"] == "5호미만제거"]
```

```python
# 5호미만제거값이 있으면 해당 행은 모두 삭제
total_df = total_df[total_df.apply(lambda row: "5호미만제거" not in row.values, axis=1)]
```

```python
total_df.shape
```

```python
total_df.head()
```

```python
# 황해북도와 제주도는 제거
total_df["시구"].unique()
```

```python
conditions = ~total_df['시구'].isin(["제주특별자치도", "황해북도"])
total_df = total_df[conditions]
```

```python
total_df.shape
```

```python
total_df["시구"].unique()
```

```python
# 계약구분 양옆 공백 제거
total_df["계약구분"] = total_df["계약구분"].map(lambda x: x.strip())
```

```python
# 시구 그룹화를 위한 매핑 딕셔너리
grouping = {
    '서울': ['인천광역시', '경기도', '서울특별시'],
    '경상도': ['경상남도', '경상북도', '대구광역시', '울산광역시', '부산광역시'],
    '충청도': ['충청남도', '충청북도', '세종특별자치시', '대전광역시'],
    '전라도': ['전라남도', '전라북도', '광주광역시', '전북특별자치도'],
    '강원도': ['강원도', '강원특별자치도']
}

# 시구를 그룹핑하여 새로운 컬럼 생성
def map_group(sigu):
    for group, regions in grouping.items():
        if sigu in regions:
            return group
    return sigu

total_df['그룹'] = total_df['시구'].apply(map_group)
```

```python
total_df.shape
```

```python
# 합계값들 전부 삭제
total_df = total_df[total_df.apply(lambda row: "합계" not in row.values, axis=1)]
```

```python
total_df.shape
```

```python
total_df.iloc[50:80, :]
```

```python
# 숫자 사이 컴마 제거
total_df["고객호수(호)"] = total_df["고객호수(호)"].map(lambda x: x.replace(",", ""))
total_df["사용량(kWh)"] = total_df["사용량(kWh)"].map(lambda x: x.replace(",", ""))
total_df["전기요금(원)"] = total_df["전기요금(원)"].map(lambda x: x.replace(",", ""))
total_df["평균판매단가(원/kWh)"] = total_df["평균판매단가(원/kWh)"].map(lambda x: x.replace(",", ""))
```

```python
# 모두 FLOAT 형으로 변환
total_df["고객호수(호)"] = total_df["고객호수(호)"].astype("float64")
total_df["사용량(kWh)"] = total_df["사용량(kWh)"].astype("float64")
total_df["전기요금(원)"] = total_df["전기요금(원)"].astype("float64")
total_df["평균판매단가(원/kWh)"] = total_df["평균판매단가(원/kWh)"].astype("float64")
```

```python
total_df["년월"][0]
```

```python
# 년월을 연도와 월로 분리
total_df['연도'] = total_df['년월'].astype(str).str[:4].astype(int)
total_df['월'] = total_df['년월'].astype(str).str[4:].astype(int)
```

```python
# 그룹 및 계약구분으로 그룹핑 후 합계 계산
grouped_df = total_df.groupby(['연도', '월', '그룹', '계약구분']).agg({
    '사용량(kWh)': 'sum',
    '전기요금(원)': 'sum',
    '평균판매단가(원/kWh)': 'mean'  # 평균으로 계산
}).reset_index()
```

### 계약구분 이름 전처리

```python
# 기존 연도와 월 데이터를 기준으로 년월 생성
grouped_df["년월"] = grouped_df["연도"].astype(str) + "-" + grouped_df["월"].astype(str).str.zfill(2)
grouped_df["년월"] = pd.to_datetime(grouped_df["년월"])
```

```python
grouped_df.head()
```

```python
# 불필요한 항목 제거
kr_power = grouped_df.drop(["연도", "월", "전기요금(원)", "평균판매단가(원/kWh)"], axis = 1)
```

```python
# 피봇 코드
kr_power.pivot(index = "년월", columns = ["그룹", "계약구분"], values = ["사용량(kWh)"]).loc[:, ("사용량(kWh)", "강원도")]
```

```python
kr_power.groupby("그룹")["계약구분"].count()
```

```python
kr_power["계약구분"].value_counts()
```

```python
# 문자열 처리 함수
def changing_name(contract_type):
    return contract_type.replace(" ", "").replace(",", "").replace("/", "").replace("및", "").replace("및", "").replace("및", "")

# '계약구분' 열에 문자열 처리 함수 적용
kr_power['계약구분'] = kr_power['계약구분'].apply(changing_name)
```

```python
kr_power["계약구분"].nunique()
```

```python
kr_power.head()
```

```python
kr_power["계약구분"].value_counts()
```

```python
# 가구내고용활동 분류 안된 행들 제거
kr_power = kr_power[~kr_power['계약구분'].str.contains('가구내고용')]
```

```python
kr_power["계약구분"].value_counts()
```

```python
# 운수업은 모두 운수창고업으로 변경
kr_power["계약구분"] = kr_power["계약구분"].apply(lambda x: "운수창고업" if x == "운수업" else x)
```

```python
# 비슷한이름끼리 묶기
kr_power["계약구분"] = kr_power["계약구분"].apply(lambda x: "하수폐기처리원료재생환경복원" if x == "수도하수폐기물처리원료재생업" else x)
kr_power["계약구분"] = kr_power["계약구분"].apply(lambda x: "정보통신업" if x == "출판영상방송통신정보서비스업" else x)
kr_power["계약구분"] = kr_power["계약구분"].apply(lambda x: "전기가스증기공기조절공급업" if x == "전기가스증기수도사업" else x)
kr_power["계약구분"] = kr_power["계약구분"].apply(lambda x: "사업시설관리사업지원서비스업" if x == "사업시설관리사업지원임대서비스업" else x)
kr_power["계약구분"] = kr_power["계약구분"].apply(lambda x: "부동산업임대업" if x == "부동산업" else x)
```

```python
# 국제외국기관은 결측치로인해 삭제
kr_power = kr_power[~kr_power['계약구분'].str.contains('국제외국기관')]
```
