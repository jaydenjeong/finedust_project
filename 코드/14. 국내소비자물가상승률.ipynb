# 데이터 불러오기

```python
cus_df = pd.read_excel("./데이터/국내 사회경제/국내 품목별 소비자물가상승률.xlsx", index_col = 0)
cus_df = cus_df.T
```

# 상승률(폭)이기 때문에 해당컬럼2번으로 컬럼추가생성

- 월이 지나면서 이전 월의 상승률 ( + )

```python
cus_df["소비자물가2"] = cus_df["소비자물가"].map(lambda x : 100 + x)
```

```python
cus_df
```

```python
for i in range(len(cus_df) - 1):
    cus_df.iloc[i + 1, -1] = round(cus_df.iloc[i, -1] + cus_df.iloc[i + 1, 0], 1)
```

```python
cus_df
```

```python
cus_df.columns
```

```python
# 나머지 컬럼도 동일하게 적용
cus_df["농축수산물"] = cus_df.iloc[:,1].map(lambda x : 100 + x)
cus_df["공업제품"] = cus_df.iloc[:,2].map(lambda x : 100 + x)
cus_df["집세"] = cus_df.iloc[:,3].map(lambda x : 100 + x)
cus_df["공공서비스"] = cus_df.iloc[:,4].map(lambda x : 100 + x)
cus_df["개인서비스"] = cus_df.iloc[:,5].map(lambda x : 100 + x)
cus_df["근원물가2"] = cus_df.iloc[:,6].map(lambda x : 100 + x)
cus_df["생활물가2"] = cus_df.iloc[:,7].map(lambda x : 100 + x)
```

```python
cus_df
```

```python
for i in range(8):
    for j in range(len(cus_df) - 1):
        cus_df.iloc[j + 1, i + 8] = round(cus_df.iloc[j, i + 8] + cus_df.iloc[j + 1, i], 1)
```

```python
cus_df
```

```python
cus_df.columns = ["소비자물가", "농축수산물", "공업제품", "집세", "공공서비스", "개인서비스", "근원물가", "생활물가",
                  "소비자물가2", "농축수산물2", "공업제품2", "집세2", "공공서비스2", "개인서비스2", "근원물가2", "생활물가2"]
```

```python
cus_df
```

```python
cus_df.reset_index(inplace = True)
```

```python
cus_df
```

```python
cus_df.rename(columns = {"index" : "일자"}, inplace = True)
```
