```python
# 우리가 사용할 5개지점으로 묶기위하여 해당지역들을 같은 value로 딕셔너리화

region_num = {"서울특별시" : 0, "경기도" : 0, "인천광역시" : 0,
              "강원도" : 1,
              "충청북도" : 2, "충청남도" : 2, "대전광역시" : 2, "세종시" : 2,
              "전라북도" : 3, "전라남도" : 3, "광주광역시" : 3,
              "경상북도" : 4, "경상남도" : 4, "부산광역시" : 4, "울산광역시" : 4, "대구광역시" : 4}
```

```python
# group컬럼으로 해당지역번호가 들어간 열 추가
df["group"] = df["시도별"].map(region_num)
```

```python
df
```

```python
# 시도별 컬럼 삭제
df.drop("시도별", axis = 1, inplace = True)
```

```python
df
```

```python
# 산업별 컬럼을 제외한 나머지 컬럼 float타입으로 변경
df.iloc[:, 1:].astype(dtype = "float64")
```

```python
# 산업별 컬럼을 제외한 나머지 컬럼 float타입으로 변경
df.iloc[:, 1:] = df.iloc[:, 1:].astype(dtype = "float64")
```

```python
merge_df = pd.merge(df.iloc[:, 0], df.iloc[:, 1:].astype(dtype = "float64"), left_index = True, right_index = True)
```

```python
merge_df
```

```python
merge_df.info()
```

```python
# 산업과 지역별로 묶기위하여 산업을 제외한 나머지 컬럼 리스트 생성
col_list = merge_df.columns[1:-1]
```

```python
group_df = pd.DataFrame()

# 지역 + 산업별로 groupby를 하며 해당컬럼은 평균으로 처리
for i in col_list:
    group_df[i] = round(merge_df.groupby(["group", "산업별"])[i].mean(), 3)
```

```python
# 날짜를 인덱스로하고 그룹을 컬럼으로 하기위해 전치
group_df = group_df.T
```

```python
group_df.info()
```

```python
group_df.reset_index(inplace = True)
```

```python
# 날짜를 인덱스로 하기위해 .삭제후 숫자만 join
group_df["index"] = group_df["index"].map(lambda x : "".join(x[:7].split(".")))
```

```python
group_df.set_index("index", inplace = True)
```

```python
group_df.columns
```

```python
region_num
```

```python
columns_list = []

# 숫자로되어있는 지역을 지역명으로 컬럼변경하기위한 튜플 리스트 생성
for i in range(5):
    if i == 0:
        i = "서울"
    elif i == 1:
        i = "강원도"
    elif i == 2:
        i = "충청도"
    elif i == 3:
        i = "전라도"
    else:
        i = "경상도"
    
    for j in ["광업", "전기업 및 가스업", "제조업", "총지수"]:
        col_name = (i, j)
        columns_list.append(col_name)
```

```python
columns_list
```

```python
# 튜플을 멀티인덱스로 생성
mtcol = pd.MultiIndex.from_tuples(columns_list)
```

```python
# 컬럼명에 멀티인덱스 적용
group_df.columns = mtcol
```

```python
group_df
```

```python
# 저장
group_df.to_csv("./국내광공업생산지수.csv")
```
