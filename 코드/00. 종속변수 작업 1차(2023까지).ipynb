```python
df_region = []

# 지역별로 돌면서 일시컬럼 자릿수 변경 및 sort_values 적용
for i in region_list:
    df = df1.loc[df1["지역"] == i]
    df["일시"] = df["일시"].map(lambda x: str(x).ljust(14, "0"))
    df = df.sort_values("일시")
    df_region.append(df)
    
# 지역별로 일자별로 정렬된 데이터프레임들 재결합
total_df = pd.concat(df_region, ignore_index = True)
total_df.head()
```

```python
# 일시를 데이트타임으로 변경하기위한 함수
def transform_datetime(data):
    data = str(data) # int가 들어가있는 경우가 있을 수 있으니 str로 변경(split을 사용하기 위해)
    data = "".join(data.split(" ")) # 띄어쓰기 삭제
    data = "".join(data.split("-")) # - 삭제
    
    # 해당 연도, 달, 일 등등을 나타내는 인덱스들을 int타입으로 변경
    year = int(data[:4])
    month = int(data[4:6])
    day = int(data[6:8])
    hour = int(data[8:10])
    minute = int(data[10:12])
    sec = int(data[12:14])
    
    # 시간이 24시가 아니라면 가져온 값으로 datetime으로 변경
    if hour != 24:
        dt = pd.to_datetime(datetime(year = year, month = month, day = day, hour = hour, minute = minute, second = sec))
        
    # 시간이 24시라면 다음날 00시이기때문에 23시로 데이터를 집어넣고 timedelta로 1시간을 추가하여 다음날로 넘어가게 코드작성
    elif hour == 24:
        dt = pd.to_datetime(datetime(year = year, month = month, day = day, hour = 23, minute = minute, second = sec)) + timedelta(hours = 1)
        
    return dt
```

```python
# 전체 일시컬럼에 함수 적용
total_df["일시"] = total_df["일시"].map(lambda x: transform_datetime(x))
```
