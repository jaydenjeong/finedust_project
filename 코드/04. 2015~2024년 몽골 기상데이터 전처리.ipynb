# 데이터 불러오기

```python
# glob 함수를 사용하여 몽골기상데이터가 들어있는 파일위치 의 csv파일만 경로이름으로 리스트 생성
filepath = glob("./데이터/몽골기상데이터/*.csv", recursive = True)
```

# 16방위 풍향 결측치 처리

- NaN값의 인덱스로 for문을 돌려 위, 아래 2개씩 4개의 값의 평균으로 대체
    - 구간이 결측일경우 결측시작의 첫번째부터 for문이 시작되므로 위 2개의 값의 평균으로 들어가게됨

```python
# 360도이므로 10을 곱하여 자리수를 맞춰줌
mongol_df["풍향"] = mongol_df["풍향"] * 10
```

```python
# 16방위 풍향이 NaN인 행의 인덱스를 추출
nan_idx = mongol_df.loc[mongol_df["풍향"].isna()].index
```

```python
mongol_df.loc[nan_idx][-30:]
```

```python
def average_angle_ignore_nan(degrees):
    """
    주어진 각도의 리스트에서 NaN 값을 무시하고 평균을 계산합니다.
    
    :param degrees: 각도의 리스트 (0-360도)
    :return: 평균 각도 (0-360도)
    """
    # NaN 값을 무시하고 유효한 각도만 선택
    valid_degrees = [deg for deg in degrees if not pd.isna(deg)]
    
    if not valid_degrees:
        return np.nan  # 유효한 각도가 없는 경우 NaN 반환
    
    # 유효한 각도를 라디안으로 변환
    radians = np.deg2rad(valid_degrees)
    
    # x, y 좌표 계산
    x_coords = np.cos(radians)
    y_coords = np.sin(radians)
    
    # x, y 좌표의 평균 계산
    x_mean = np.mean(x_coords)
    y_mean = np.mean(y_coords)
    
    # 평균 좌표를 각도로 변환
    mean_rad = np.arctan2(y_mean, x_mean)
    mean_deg = np.rad2deg(mean_rad)
    
    # 결과를 0-360도 사이의 값으로 변환
    mean_deg = mean_deg % 360
    if mean_deg > 360:
        mean_deg = mean_deg - 360
    
    return round(mean_deg, -1)
```

```python
# NaN값의 첫번째 인덱스부터 for문을 돌며 결측치 처리
for i in nan_idx:
    mongol_df.iloc[i, 3] = round(average_angle_ignore_nan(mongol_df.iloc[i-2:i+3, 3]),-1)
```

# 피봇테이블로 변경

```python
pivot_df = mongol_df.pivot_table(index = "일시", columns = "지점", values = mongol_df.columns[2:])
```

```python
pivot_df
```

```python
pd.get_dummies(pivot_df["풍향"])
```

```python
dummie_list = []

for loc in pivot_df["풍향"].columns:
    dum = pd.get_dummies(pivot_df["풍향"][loc])
    dum.columns = [(loc, i) for i in dum.columns]
    dummie_list.append(dum)
```

```python
dummie_list[0]
```

```python
dum_df = pd.concat(dummie_list, axis = 1)
dum_df.columns = [("풍향", i) for i in dum_df.columns]
```

```python
dum_df = dum_df.applymap(lambda x : x * 1)
```

```python
dum_df
```

```python
pivot_df = pivot_df.drop("풍향", axis = 1)
```

```python
total_df = pd.concat([pivot_df, dum_df], axis = 1)
```
