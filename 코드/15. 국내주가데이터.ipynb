```python
import pandas as pd
import numpy as np
```

# 국내 코스닥 지수

```python
kosdaq_df = pd.read_csv("../데이터/국내 주가 지수/코스닥.csv")
kosdaq_df
```

```python
kosdaq_df.info()
```

```python
# 날자컬럼의 띄어쓰기 삭제
kosdaq_df["날짜"] = kosdaq_df["날짜"].map(lambda x : "".join(x.split(" ")))
```

```python
# 4개의 컬럼 float타입으로 변경 -> , 삭제 및 타입변경
kosdaq_df[["종가", "시가", "고가", "저가"]] = kosdaq_df[["종가", "시가", "고가", "저가"]].applymap(lambda x : float("".join(x.split(","))))
```

```python
# float타입으로 변경하기위해 거래량의 단위인 K와 B를 해당 값을 곱하여 변경
kosdaq_df["거래량"] = kosdaq_df["거래량"].map(lambda x : (float(x[:-1]) * 1000 if x[-1] == "K" 
                                                                                    else float(x[:-1]) * 100000000 
                                                                                if x[-1] == "B"
                                                                                    else float(x[:-1]) * 1000000))
```

```python
# 변동은 퍼센트이므로 %를 제거한 뒤 100을 곱하여 float타입으로 변경
kosdaq_df["변동 %"] = kosdaq_df["변동 %"].map(lambda x : float(x[:-1]) * 100)
```

```python
# 날짜와 변동%의 컬럼명 변경
kosdaq_df.rename(columns = {"날짜" : "일자", "변동 %" : "변동"}, inplace = True)
```

```python
# 일자를기준으로  오름차순으로 정렬
kosdaq_df.sort_values("일자", ignore_index = True, inplace = True)
```

# 국내 코스피 지수

```python
kospi_df = pd.read_csv("../데이터/국내 주가 지수/코스피.csv")
kospi_df
```

```python
kospi_df.info()
```

```python
kospi_df["날짜"] = kospi_df["날짜"].map(lambda x : "".join(x.split(" ")))
```

```python
kospi_df[["종가", "시가", "고가", "저가"]] = kospi_df[["종가", "시가", "고가", "저가"]].applymap(lambda x : float("".join(x.split(","))))
```
