## 독립변수 읽기

```python
# 각 기상 데이터 읽기
kr_df = pd.read_csv("./독립변수/국내기상데이터.csv", index_col = 0, header = [0, 1])
cn_df = pd.read_csv("./독립변수/중국기상데이터.csv", index_col = 0, header = [0, 1])
sea_df = pd.read_csv("./독립변수/해양기상데이터.csv", index_col = 0, header = [0, 1])
mg_df = pd.read_csv("./독립변수/몽골기상데이터.csv", index_col = 0, header = [0, 1])
cn_dust = pd.read_csv("./독립변수/중국미세먼지.csv", index_col = 0, header = [0, 1])
```

```python
# kr_df에 시간/월 범주 데이터가 있던 것 제거
kr_df = kr_df.iloc[:, 38:]
```

### 독립변수 통합

```python
# 한국기상, 중국기상, 해양기상, 몽골기상 데이터 concat
indep_df = pd.concat([kr_df, cn_df, sea_df, mg_df], axis = 1)
```

```python
# 병합된 데이터프레임의 인덱스를 datetime 타입으로 변경
indep_df.index = indep_df.index.map(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
```

```python
# 미세먼지 데이터가 입력될 컬럼 생성
for column in cn_dust.columns:
    indep_df[column] = np.nan
```

```python
# 각 날짜별로 미세먼지 농도 값 입력
for i in range(len(indep_df)):
    # 미세먼지 데이터 특정을 위한 날짜값
    date = indep_df.index.date[i]
    # concat_df에 입력할 위치를 특정하기 위한 datetime값
    datetime = indep_df.index[i]
    # 데이터 특정, 입력 위치 특정을 위한 컬럼명 목록
    column = cn_dust.columns
    # 데이터 입력
    indep_df.loc[datetime, column] = cn_dust.loc[str(date), :]
```

```python
# 월 / 시간 값을 범주화한 데이터프레임 생성
datetime_df = pd.concat([pd.get_dummies(indep_df.index.month, prefix = "month"), 
                         pd.get_dummies(indep_df.index.hour, prefix = "hour")], axis = 1)
```

```python
# 인덱스값을 일시로 변경
datetime_df.index = indep_df.index
```

```python
# concat_df와 병합
indep_df = pd.concat([indep_df, datetime_df], axis = 1)
```

### 종속변수 병합

- pm10의 경우 15년 1월 1일부터, pm2.5의 경우 19년 1월 1일부터 데이터가 시작됨
- 기간이 적절한 2개의 데이터프레임을 생성하여 별도로 저장하는 방향으로 진행

```python
# 병합을 위해 인덱스값을 datetime 타입으로 변경
pm10_df.index = pd.to_datetime(pm10_df.index)
pm25_df.index = pd.to_datetime(pm25_df.index)
```

```python
# pm10 최종본 데이터 생성
pm10_final = pd.concat([pm10_df, indep_df], axis = 1)
```
