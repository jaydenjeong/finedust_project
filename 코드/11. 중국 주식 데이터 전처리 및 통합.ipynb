### 결측일 확인 및 불필요한 날짜 제거

```python
# cn_a50에는 있으나 cn_sha는 없는 날짜 확인
cn_a50[~cn_a50["날짜"].isin(cn_sha["날짜"])]
```

```python
# cn_a50에는 있으나 cn_szse에는 없는 날짜 확인
cn_a50[~cn_a50["날짜"].isin(cn_szse["날짜"])]
```

```python
# 결측된 두 날짜의 리스트가 서로 동일하다는 것을 확인
cn_a50[~cn_a50["날짜"].isin(cn_sha["날짜"])].index == cn_a50[~cn_a50["날짜"].isin(cn_szse["날짜"])].index
```

```python
# 결측 날짜의 길이 확인
len(cn_a50[~cn_a50["날짜"].isin(cn_sha["날짜"])])
```

- 결측일 대부분의 변동률이 0%인 것으로 확인됨
- 세부조사 결과 변동률이 0%인 날은 주말로 불필요한 데이터
- 단 변동률이 있는 2일은 cn_sha와 cn_szse측 결측으로 생각됨
    - interpolate를 사용하여 처리 예정

```python
# 결측일 중 변동률이 0인 날짜들의 인덱스 특정
drop_idx = cn_a50[(~cn_a50["날짜"].isin(cn_sha["날짜"])) & (cn_a50["변동 %"] == 0)].index
```

```python
# drop_idx를 cn_a50에서 제거
cn_a50 = cn_a50.drop(drop_idx, axis = 0).reset_index(drop = True)
```

### 병합 작업

```python
# 각 데이터프레임에서 날짜와 종가를 제외한 나머지 피처 제거
cn_a50 = cn_a50.loc[:, ["날짜", "종가"]]
cn_szse = cn_szse.loc[:, ["날짜", "종가"]]
cn_sha = cn_sha.loc[:, ["날짜", "종가"]]
```

```python
# 각 종가 피처의 명칭을 데이터프레임의 명칭으로 변경
cn_a50.columns = ["날짜", "cn_a50"]
cn_szse.columns = ["날짜", "cn_szse"]
cn_sha.columns = ["날짜", "cn_sha"]
```

```python
# cn_stock이라는 데이터프레임으로 병합
cn_stock = pd.merge(cn_a50, cn_szse, on = "날짜", how = "outer")
cn_stock = pd.merge(cn_stock, cn_sha, on = "날짜", how = "outer" )
```

### 결측치 제거

```python
# 결측치 확인
cn_stock.info()
```

- cn_szse와 cs_sha에 2개씩의 결측이 있는 것이 확인됨

```python
# 2개의 결측 데이터 확인
cn_stock[cn_stock["cn_szse"].isna()]
```

```python
# interpolate("linear")를 활용하여 2개의 결측치 제거
cn_stock = cn_stock.interpolate("linear")
```

### 주말, 공휴일 추가

- 주말과 공휴일엔 거래가 없는 주식 특성 상 해당 데이터는 자연스럽게 결측되어 있음
    - 15년 1월 1일부터 24년 5월 19일까지는 3427일로 약 1/3의 데이터가 결측된 상태임
    
- 주말과 공휴일의 경우 ffill()을 활용해 처리하기로 한다

```python
# 날짜 피처를 datetime으로 변환
cn_stock["날짜"] = pd.to_datetime(cn_stock["날짜"])
```

```python
# 날짜 피처를 인덱스로 설정
cn_stock = cn_stock.set_index("날짜")
```

```python
# 모든 날짜를 보유한 리스트 생성
date_list = pd.date_range(start = "2015-01-01", end = "2024-05-19")
```

```python
# cn_stock에 없는 일시에 대해서만 np.nan 값을 지니는 데이터 생성
for date in date_list:
    if date not in cn_stock.index:
        cn_stock.loc[date, :] = np.nan
```

```python
# 날짜(index)를 기준으로 정렬
cn_stock = cn_stock.sort_index()
```

```python
# ffill 기법으로 결측치 처리
cn_stock = cn_stock.ffill()
```

```python
cn_stock
```

- ffill로는 1월 1일부터 1월 4일까지의 데이터를 처리할 수 없음
- bfill로 4일간의 데이터 처리

```python
# bfill 적용
cn_stock = cn_stock.bfill()
```
