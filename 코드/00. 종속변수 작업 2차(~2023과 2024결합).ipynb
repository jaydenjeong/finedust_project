```python
# 24년 미세먼지데이터가 있는 경로 glob 를 사용하여 경로 리스트

filepath = glob("./데이터/국내 초미세먼지 2024년/*/*.xls", recursive = True)
```

```python
filepath
```

```python
dfs = []

# 데이터 불러오기 확인을 하여 불러올 방법을 for문을 통하여 각 경로의 파일마다 적용
for filename in filepath:
    df = pd.read_excel(filename, skiprows = 3)
    df.drop([0, 1], axis = 0, inplace = True)
    df.drop(df.columns[5:], axis = 1, inplace = True)
    df.drop(["PM10", "PM2.5"], axis = 1, inplace = True)
    df.columns = ["날짜", "PM10", "PM25"]
    
    # 지역명 컬럼을 해당 파일의 지역이름으로 생성
    df["지역"] = filename.split("\\")[-2]
    # 컬럼순서 변경
    df = df.loc[:, ["지역", "날짜", "PM10", "PM25"]]
    # 컬럼명 변경
    df.columns = ["지역", "일시", "PM10", "PM25"]
    dfs.append(df)
    
# 24년 데이터 concat
df2024 = pd.concat(dfs, ignore_index = True)
df2024.head()
```

```python
df2024.info()
```

```python
# 일시를 데이트타임으로 변경하기위한 함수
def transform_datetime(data):
    data = str(data)
    data = "".join(data.split(":"))
    data = "".join(data.split("-"))
    
    year = int(data[:4])
    month = int(data[4:6])
    day = int(data[6:8])
    hour = int(data[8:10])
    minute = 0
    sec = 0
    
    if hour != 24:
        dt = pd.to_datetime(datetime(year = year, month = month, day = day, hour = hour, minute = minute, second = sec))
    elif hour == 24:
        dt = pd.to_datetime(datetime(year = year, month = month, day = day, hour = 23, minute = minute, second = sec)) + timedelta(hours = 1)
        
    return dt
```

```python
df_region = []

# 지역별로 돌며 일시컬럼 datetime으로 변경하기 위하여 ljust 적용
for i in df2024["지역"].unique():
    df = df2024.loc[df2024["지역"] == i]
    df["일시"] = df["일시"].map(lambda x: str(x).ljust(14, "0")) # 14자리가 되지않는 문자열 오른쪽에 0으로 채우기
    df = df.sort_values("일시") # 오름차순으로 정렬
    df_region.append(df)
    
# 정렬된 데이터프레임 결합
df2024 = pd.concat(df_region, ignore_index = True)
df2024.head()
```

```python
# 일시컬럼 datetime으로 변환하기위해 함수 적용
df2024["일시"] = df2024["일시"].map(lambda x : transform_datetime(x))
```

```python
# timedelta 함수를 사용하여 요일과시간에따라 묶을 구룹 컬럼 정리
time_list = [df_region2[0].loc[0, "일시"] + (timedelta(hours = 1) * i) for i in range(0, 82440)]
time_df = pd.DataFrame(time_list)
time_df.columns = ["일시"]
time_df["group"] = (time_df.index) // 3
time_df
```

```python
concat_list = []

# 지역별로 time 데이터프레임과 결합
for dfi in df_region2:
    concat_dfi = pd.merge(time_df, dfi, how = "left", on = "일시")
    concat_list.append(concat_dfi)
    
# 결합된 데이터프레임들 concat
concat_df = pd.concat(concat_list, ignore_index = True)
```

```python
# 종속변수가 될 pm10과 pm25의 결측치가 - 로 되어있어 float타입으로 변경이 안되기 때문에
# isinstance 와 isnumeric을 사용하여 숫자라면 float타입으로 변경 아니라면 nan으로 변경

concat_df["PM10"] = concat_df["PM10"].map(lambda x: float(x) if isinstance(x, str) and x.isnumeric() else x)
concat_df["PM25"] = concat_df["PM25"].map(lambda x: float(x) if isinstance(x, str) and x.isnumeric() else x)
```

```python
# 연속된 True 값을 계산하는 함수
def calculate_consecutive_trues(series):
    n = len(series)
    result = [0] * n
    count = 0

    # 첫 번째 패스: 연속된 True의 그룹 길이를 계산
    for i in range(n):
        if series[i]:
            count += 1
        else:
            count = 0
        result[i] = count

    # 두 번째 패스: 그룹의 마지막 True 값에 그룹의 전체 길이를 설정
    final_result = [0] * n
    i = 0
    while i < n:
        if result[i] > 0:
            length = result[i]
            for j in range(length):
                final_result[i - j] = length
            i += length
        else:
            i += 1

    return final_result
```
